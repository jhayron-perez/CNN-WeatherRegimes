```python
import numpy as np
import xarray as xr
import pandas as pd
import datetime as dt
import glob
import os
```

# Resample Z500

```python
path_files =  f'/glade/work/jhayron/Weather_Regimes/ERA5/PL/'
files_geop = glob.glob(f'{path_files}ERA5_PL_GEOP500_*.nc')
path_output_daily =  f'/glade/work/jhayron/Weather_Regimes/ERA5/Daily/Z500/'
```

```python
for file in files_geop:
    print(file)
    dataset_temp_hourly = xr.open_dataset(file)
    dates_hourly = pd.to_datetime(dataset_temp_hourly.time).date
    dates_daily = np.unique(dates_hourly)

    for date in dates_daily:
        dataset_temp_daily = dataset_temp_hourly.isel(time=np.where(dates_hourly == date)[0]).mean('time')
        date_str = str(date).replace('-','_')
        dataset_temp_daily.to_netcdf(f'{path_output_daily}Z500_{date_str}.nc')
    os.remove(file)
```
